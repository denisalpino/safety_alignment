```python
from pathlib import Path
import sys

project_root = str(Path().resolve().parent)
sys.path.insert(0, project_root)
```

```python
from datasets import load_dataset

from src.vis.eda import (
    create_prompt_length_distribution,
    create_length_heatmap,
    create_grouped_bar_chart
)
```

```python
dataset = load_dataset("Galtea-AI/galtea-red-teaming-clustered-data")["train"].to_pandas()
print(dataset.shape)
dataset.head()
```

```python
CLUSTERS = {
    0: "Ambiguous Requests",
    1: "Jailbreak & Roleplay Attacks",
    2: "Financial Stuff & Fraud",
    3: "Toxicity & Hate Speech",
    4: "Violence & Illegal Activities",
    5: "Privacy Violations",
}

dataset.cluster = dataset.cluster.map(CLUSTERS)
```

```python
create_grouped_bar_chart(dataset)
```

```python
create_prompt_length_distribution(dataset)
```

```python
create_length_heatmap(dataset)
```

```python
dataset[dataset.prompt_length <= 15].sort_values("prompt_length")
```

```python
dataset = dataset[dataset.prompt_length >= 15].reset_index(drop=True).drop(columns=["prompt_length"])
dataset.cluster = dataset.cluster.map({v: k for k, v in CLUSTERS.items()})
dataset.shape
```

```python
dataset = dataset.sample(n=10_000, random_state=42, ignore_index=True)
```

```python
dataset.to_csv(f"{project_root}/data/dataset.csv")
```
